```python
import pandas as pd

file_path = r"C:\Users\Aakriti K C\Desktop\DA and AI\project 1 IMDB\bda66-imdb-movie-reviews-scraping-cleaning-the-movie-ratings-bda66imdb-s526124\most_popular_movies.csv"
df = pd.read_csv(file_path)
```

```python
# Info about the dataset
print("Dataset Info Before Cleaning:")
print(df.info())
print("\nSample Data:")
print(df.head())
```

```python
# Handling missing values
# Replacing "N/A" and empty strings with NaN
df.replace("N/A", pd.NA)

# Filling missing values in specific columns
df['Description'] = df['Description'].fillna("No description available")
df['Genre'] = df['Genre'].fillna("Unknown")

# Dropping rows where critical columns are missing
df.dropna(subset=['Title', 'Best Rating', 'Worst Rating','Rating Value'])

# Removing duplicates
df.drop_duplicates(subset = ['Title'])
```

```python
# Info after cleaning
print("\nDataset Info After Cleaning:")
print(df.info())

# Saving the cleaned data
cleaned_file_path = r"C:\Users\Aakriti K C\Desktop\DA and AI\project 1 IMDB\bda66-imdb-movie-reviews-scraping-cleaning-the-movie-ratings-bda66imdb-s526124\cleaned_most_popular_movies2.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")
```
